# Constructor de la clase paciente

```python
#creacion de la clase
class paciente:
    
    #constructor
    def __init__(self, nombre, tipo_documento, documento, sexo, dia, mes, anio, cama):
        self.nombre = nombre
        self.tipo_documento = tipo_documento
        self.documento = documento
        self.sexo = sexo
        self.dia = dia
        self.mes = mes
        self.anio = anio
        self.medicamentos = {}  
        self.signos_vitales = {}
        self.cama = cama
        self.examenes = {}
        
    
    #funcion para definir la fecha de nacimiento
    def fecha_nacimiento(self):
        return(f"{self.dia}/{self.mes}/{self.anio}")
    
    #funcion para insertar los medicamentos del paciente
    def recetar_medicamento(self):
        while True: #uso del while para que la funcion se repita las veces que desee el usuario
            print("\nIngrese el medicamento usando este formato --> nombre del medicamento: Dosis ")
            a=input("\ningrese Infomación ")#variable usada para guardar la informacion suministrada
            
            try: #uso del try - except para validar la entrada de los datos
                medicamento,dosis=a.split(":")
            except:
                a=input("\ningrese información Por favor siga el formato MEDICAMENTO: VALOR ")
                medicamento,dosis=a.split(":")
            self.medicamentos[medicamento]=dosis#almacenamiento de la informacion suministrada
            continuar=input("\nDesea ingresar otro dato? S para continuar ")#se le pregunta al usuario si desea continuar
            if continuar.upper()!="S":#validacion de la opcion ingresada por el usuario
                break
    
    #funcion que registra las imagenes diagnosticas del paciente
    def imagenes_diagnosticas(self):
        self.imagen = input("Inserte resultados de las imagenes diagnosticas del paciente (descripcion) ")
        return self.imagen
            
    def registrar_signos_vitales(self):
        while True:#uso del while para que la funcion se repita las veces que desee el usuario
            print("\nIngrese los signos vitales usando este formato --> Signo vital:Valor ")
            a=input("\ningrese información")#variable usada para guardar la informacion suministrada
            try: #uso del try - except para validar la entrada de los datos
                signo,valor=a.split(":")
            except:
                a=input("\ningrese información Por favor siga el formato SIGNO VITAL: VALOR ")
                signo,valor=a.split(":")
            self.signos_vitales[signo]=valor#almacenamiento de la informacion suministrada
            continuar=input("\nDesea ingresar otro dato? S para continuar ")#se le pregunta al usuario si desea continuar
            if continuar.upper()!="S":#validacion de la opcion ingresada por el usuario
                break
     
    #funcion que registre los examenes del paciente
    def resultados_examen(self):
        while True:#uso del while para que la funcion se repita las veces que desee el usuario
            print("\nIngrese los resultados usando este formato --> examen:resultado ")
            a=input("ingrese información ")#variable usada para guardar la informacion suministrada
            try: #uso del try - except para validar la entrada de los datos
                examen,resultado=a.split(":")
            except:
                a=input("\ningrese información Por favor siga el formato EXAMEN:RESULTADO ")
                examen,resultado=a.split(":")
            self.examenes[examen]=resultado#almacenamiento de la informacion suministrada
            print(examen,resultado,self.examenes)
            continuar=input("\nDesea ingresar otro dato? S para continuar ")#se le pregunta al usuario si desea continuar
            if continuar.upper()!="S":#validacion de la opcion ingresada por el usuario
                break  
                
    def mostrar_examenes(self):
        print("\n\tINFORMACION DEL PACIENTE")
        print(f"\nNombre: {self.nombre}                          Identificacion {self.tipo_documento} {self.documento}")
        print(f"\nFecha de nacimiento: {self.fecha_nacimiento()}          Genero:{self.sexo}")
        print(f"\nCama:{self.cama}")
        print("---------------------------------------------------------------------------")
        print("\nExamenes                           Resultado")
        for x in self.examenes:
            print(f'\n{x}                           {self.examenes[x]}')
        
def entrada_datos():
    #insercion de los datos del paciente
        nombre = input("\nInserte nombre del paciente ")
        nombre = nombre.capitalize() #.capitalize sirve para que la primera letra este en mayuscula
        tipo_documento = input("\nInserte tipo de documento")
        documento = int(input("\nInserte documento del paciente "))
        dia = int(input("\nInserte dia de nacimiento "))
        while (dia < 1 or dia > 31):#validacion de la variable dia: no menor a 1 o no menor a 31
            dia = int(input("\nInserte dia de nacimiento valido!! "))
        
        mes = int(input("\nInserte mes de nacimiento "))
        while (mes < 1 or mes > 12):#validacion de la variable mes: no menor a 1 o no menor a 12
            mes = int(input("\nInserte año de nacimiento valido!! "))
            
        anio = int(input("\nInserte año de nacimiento "))
        while (anio < 1930 or anio > 2023):#validacion de la variable año: no menor a 1930 o no menor a 2023
            anio = int(input("\nInserte año de nacimiento valido!! "))
        
        sexo = (input("\nInserte el sexo del paciente F = Femenino M = masculino "))
        while (sexo.upper() != "F" and sexo.lower() != "m"):#validacion de la variable sexo para solo insertar f y m
             sexo = input("\nInserte un sexo valido -> F = Femenino M = masculino ")
        
        cama = int(input("\nInserte el numero de cama del paciente (1 - 300) "))
        while (cama < 1 or cama > 300):#validacion de la variable cama: no menor a 1 o no menor a 300
            cama = int(input("\nInserte numero de cama valido!! "))
        pacientes = paciente(nombre, tipo_documento, documento, sexo, dia, mes, anio, cama)#almacenamiento de los datos del paciente en el objeto pacientes
    
    
```
